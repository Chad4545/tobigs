```python
def prob(n):
    '''
    A = 30m 간 횟수
    B = 15m 간 횟수
    type_120: n-1 번째 120m 가는 경우의 수
    type_135: n-1 번째 135m 가는 경우의 수
    type_120_prob: n-1 번째 120m 가는 확률
    type_135_prob: n-1 번째 135m 가는 확률
    prob_result = result
    '''
    prob_result = 0
#n=6
    if n % 2 == 0:
        
        type_120=[]
        type_135=[]
        type_120_prob=0
        type_135_prob=0
        
        # 1, 2, 3, 4, 5
        for idx in range(n):
            A = idx
            B = n -1 -idx
            
            if A*30 + B*15 == 120:
                type_120.append(A)
                type_120.append(B)
            
            elif A*30 + B*15 == 135:
                type_135.append(A)
                type_135.append(B)
                
        type_120_prob = math.factorial(n-1)/math.factorial(type_120[0])/math.factorial(type_120[1])*(0.5)**(n-1)
        type_135_prob = math.factorial(n-1)/math.factorial(type_135[0])/math.factorial(type_135[1])*(0.5)**(n-1)
        
        prob_result = 0.5 * type_120_prob + type_135_prob
        return prob_result
                
    
    else:
        type_120=[]
        type_135=[]
        type_120_prob=0
        type_135_prob=0
        
        # 1, 2, 3, 4, 5
        for idx in range(n):
            A = idx
            B = n -1 -idx
            
            if A*30 + B*15 == 120:
                type_120.append(A)
                type_120.append(B)
            
            elif A*30 + B*15 == 135:
                type_135.append(A)
                type_135.append(B)
                
        type_120_prob = math.factorial(n-1)/math.factorial(type_120[0])/math.factorial(type_120[1])*(0.5)**(n-1)
        type_135_prob = math.factorial(n-1)/math.factorial(type_135[0])/math.factorial(type_135[1])*(0.5)**(n-1)
        
        prob_result = 0.5 * type_120_prob + type_135_prob
        
        return prob_result
        
        
        
```

### 단어로 이루어진 txt파일을 읽어와 반복 스펠링 단어의 개수를 출력하는 프로그램을 작성하시오 (첨부 txt파일 확인)

```python
import pandas as pd
data = pd.read_csv('a_dic.txt',sep=' ', header=None)
data.columns = ["word"]
```

```python
#단어단위로 iterate
count = 0
for word in data['word']:
    
    # 각 단어를 dict으로 만듦
    # alpha -> {'a':2,'l':1,'p':1,'h':1}
    word = word.lower()
    check_dict = {}
    for char in word:
        check_dict[char] = check_dict.get(char,0) +1
    
    # 단어의 길이와 check_dict의 길이가 같다면
    # 유니크한 char조합의 단어라는 뜻이므로 반복스펠링단어
    if len(check_dict.keys()) == len(word):
        count += 1
        print('----')
        print('check_word: ', word)
        print(word,': repeat spelling word ')
    
    # 그 외는 char가 하나이상인 경우가 있으므로, 그 char가 연속인지 체크
    else:
        # 2개 이상 char를 담을 리스트 
        check_char=[]
        for key, val in check_dict.items():
            if val > 1:
                check_char.append(key)
        print('----')
        print('check_word: ', word)
        print('check_char_in_word: ', check_char)
        
        # check해야할 char를 얻었다.b
        # word의 index를 활용하여 체크하자
        
        second_count = len(check_char)
        #print('len of secound_count: ',second_count)
        
        for char_in_check in check_char:
            print('char_in_check: ', char_in_check)
            # alpha의 경우 a의 인덱스가 담긴 리스트 [0,4]
            # banana의 경우 a의 인덱스가 담긴 리스트 [1,3,5]
            last_check_list=[idx for idx, char in enumerate(word) if char == char_in_check]

            print('last_check_list',last_check_list)
            if max(last_check_list) - min(last_check_list) == len(last_check_list) - 1:
                second_count -= 1
           
        if second_count == 0:
            count += 1
            print(word,': repeat spelling word ')
        else:
            print(word,': not repeat spelling word')
                
print(count)

            
            
                
                
    
        
        
```

- 6은 1,2,3,6의 4개의 약수를 가지고 8은 1,2,4,8의 4개의 약수를 가지며,10은 1,2,5,10의 4개의 약수를 가진다



- 따라서 5이상 10이하의 범위에서 4개의 약수를 가지는 수는 6,8,10의 3개이다.


- 이렇게 약수의 개수 k와 수의 범위a,b(0<=a,b<=100000)가 주어질 때 조건을 만족하는 수와 개수를 출력하는 프로그램을 작성하라

```python
k = int(input('약수의 개수: '))
a = int(input('시작 범위: '))
b = int(input('끝 범위: '))
answer_numbers = []

# 약수의 개수가 홀수라는건 제곱수
if k%2 == 1:

    # a에서 b까지 정수가 돌아갑니다.
    for num_a in range(a,b+1):
        
        #k와 비교할 count
        count_a=0
        #print('----------')
        #print('Number: ',num_a)
        
        #num을 1부터 num까지의 숫자로 나눠보자 
        for i in range(1,num_a+1):
            if num_a%i == 0:
                count_a += 1
        
        if count_a == k:
            answer_numbers.append(num_a)
            
# 약수의 개수가 짝수            
else:
    
    # a에서 b까지 정수가 돌아갑니다.
    for num_b in range(a,b+1):
        
        #k와 비교할 count
        count_b = 0
        #print('---------')
        #print('Number: ',num_b)
        
        #num을 1부터 num까지의 숫자로 나눠보자
        for j in range(1,num_b+1):
            if num_b%j == 0:
                count_b +=1
            
            if count_b ==k:
                answer_numbers.append(num_b)
            
print(answer_numbers)            
print(len(answer_numbers),' 개 입니다.')
    
        
    
```

# 5.  

- B(2<=B<=)를 입력했을때 진수대칭변환을 찾고 싶다.

- 진수대칭변환일아 1부터 300까지의 수 중에서 그 제곱을 B진수로 표현했을 때,

- 좌우대칭이 되는 경우를 말한다.

- B 입력시 진수대칭변환 출력 후 개수를 구하는 프로그램을 작성하라.

```python
def convert(n, B):
    T = "0123456789"
    quotient, remainder = divmod(n, B)
    
    # 몫이 0이면  
    if quotient == 0:
        return T[remainder]
    else:
        return convert(quotient, B) + T[remainder]
```

```python
def symmetry(a):
    lenn = len(a)
    
    #변환한 수가 홀수
    if lenn%2 == 1:
        idx=int(lenn/2)
        forward = a[:idx]
        backward = a[idx+1:]
        backward = backward[::-1]
        return forward == backward
    
    #변환한 수가 짝수
    else:
        idx = int(lenn/2)
        forward = a[:idx]
        backward = a[idx:]
        backward = backward[::-1]
        return forward == backward
        
        
```

```python
B = int(input('B진수로 변환: '))
count = 0
for i in range(1,301):
    square=i*i
    check=convert(square,B)
    
    if symmetry(check):
        count+=1
        print(i,'-> 제곱: ', square,'-> 대칭변환:  ', check)
print(count,'개의 진수대칭변환이 있습니다.')
    
    
```
